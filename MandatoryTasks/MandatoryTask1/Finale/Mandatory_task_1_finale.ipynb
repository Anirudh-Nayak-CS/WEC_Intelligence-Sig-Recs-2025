```python
## from part C fine tuned model

device="cuda" if torch.cuda.is_available() else "cpu"
from transformers import pipeline,AutoTokenizer
from transformers import DistilBertTokenizerFast

tokenizer = AutoTokenizer.from_pretrained("DingyNayak/Roberta")
pipe = pipeline("question-answering", model="DingyNayak/Roberta",tokenizer=tokenizer,device=device)


```

```python
## from part B

import torch
import torch.optim as optim
from transformers import AutoModelForSeq2SeqLM, T5Tokenizer,Seq2SeqTrainingArguments,DataCollatorForSeq2Seq,Seq2SeqTrainer

pipeline2 = pipeline(
    task="text2text-generation",
    model="google-t5/t5-base",
    dtype=torch.float16,
    device=0
)
```

## Demonstration




#### Question 1


```python
question="What fell on Newton's head?"
context="An apple fell on Newton's head."
answer="Apple."
```

```python
# using the fine tuned model to get ans

prediction=pipe(question=question,context=context)

labels=""

print(f"question: {question}")
print(f"context: {context}")

print(f"Received prediction from fine tuned model: {prediction['answer']}")
print(f"Actual answer: {answer}")

# using t5 model to get french conversion
translated_pred_in_french=pipeline2("Translate English to French:"+prediction['answer'])
expected_ans="Pomme"

print(f"Answer in French: {translated_pred_in_french[0]['generated_text']} ")
print(f"Expected answer in French : {expected_ans}")
```
